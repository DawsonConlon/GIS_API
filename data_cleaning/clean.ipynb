I want to clean the sask df I have to take the sensor dict column and clean it so its easy to analyze

```python
import pandas as pd
import geopandas as gpd
import ast # used to convert string representation of list to actual list
```

```python
# import api1/saskatoon_sensors.csv
df = pd.read_csv("/Users/dawsonconlon/Coding/Python/GIS_API/api1/saskatoon_sensors.csv")
print(df.head())
```

```python
df["sensors"] = df["sensors"].apply(ast.literal_eval)
```

```python
rows = []

for _, row in df.iterrows():
    for sensor in row["sensors"]:
        rows.append({
            "station_id": row["id"],
            "station_name": row["name"],
            "locality": row["locality"],
            "latitude": row["coordinates.latitude"],
            "longitude": row["coordinates.longitude"],
            "pollutant": sensor["parameter"]["displayName"],
            "parameter": sensor["parameter"]["name"],
            "units": sensor["parameter"]["units"]
        })

clean_df = pd.DataFrame(rows)
```
