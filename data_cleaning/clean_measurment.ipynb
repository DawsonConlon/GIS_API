This notebook will clean the df we pulled from the api and make it rready for analysis

```python
# import libraries
import requests
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_csv("/Users/dawsonconlon/Coding/Python/GIS_API/api1/measurements.csv")
```

```python
def clean_air_quality_data(df):
    
    """"
    # Function to clean air quality measurement data from OpenAQ API
    
    # Parameters:
    # df (pd.DataFrame): Raw measurement data from OpenAQ API
    
    # Returns:
    # pd.DataFrame: Cleaned measurement data with relevant columns, proper data types, and sensor coordinates
    
    
    """
    
    # keep only useful columns
    df = df[
        [
            "period.datetimeFrom.utc",
            "parameter.name",
            "parameter.units",
            "value"
        ]
    ]

    # rename columns
    df = df.rename(columns={
        "period.datetimeFrom.utc": "datetime",
        "parameter.name": "pollutant",
        "parameter.units": "units",
        "value": "measurement"
    })

    # convert datetime
    df["datetime"] = pd.to_datetime(df["datetime"])

    # add sensor coordinates
    df["latitude"] = 52.136055
    df["longitude"] = -106.662796

    # sort by time
    df = df.sort_values("datetime")

    # drop missing measurements
    df = df.dropna(subset=["measurement"])

    return df
```

```python
clean_df = clean_air_quality_data(df)

print(clean_df.head())
```

```python
# save cleaned data
clean_df.to_csv("clean_measurements.csv", index=False)
```
